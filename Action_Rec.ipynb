```
import cv2
import os
import numpy as np
from PIL import Image
from google.colab.patches import cv2_imshow
onlyfiles = onlyfiles = ['golf','ride_horse','pushup','dive','kiss']
count = 1
final =[]
target =[]
for file in onlyfiles:
  folder = "/content/gdrive/My Drive/Colab Notebooks/Action Recognition/Action/" + file 
  #imgfolder = folder+"/"+ str(count)
  
  for di in os.listdir(folder):
    image=[]
    for i in range(0,8):
      imgpath = "/content/gdrive/My Drive/Colab Notebooks/Action Recognition/Action/" + file+"/"+di+"/frame"+str(i*5)+".jpg"
      print(imgpath)
      img = cv2.imread(imgpath)
      #cv2_imshow(img)
      #img.show()
      img = cv2.resize(img,(128,128))
      img = np.array(img)
      #print(img.shape)
      image.append(img)
      
    final.append(image)
    count = count + 1
    if( file == 'golf'):
      t = np.array([1,0,0,0,0])
      target.append(t)
    elif( file == 'ride_horse'):
      t = np.array([0,1,0,0,0])
      target.append(t)
    elif( file == 'pushup'):
      t = np.array([0,0,1,0,0])
      target.append(t)
    elif( file == 'dive'):
      t = np.array([0,0,0,1,0])
      target.append(t)
    elif( file == 'kiss'):
      t = np.array([0,0,0,0,1])
      target.append(t)

    if(file == "golf" and count == 106):
      continue
    elif(file == "ride_horse" and count == 117):
      continue
    elif(file == "pushup" and count == 104):
      continue
    elif(file == "dive" and count == 128):
      continue
    elif(file == "kiss" and count == 103):
      continue


  
```

```
 final = np.array(final)
```

```
final.shape
```

```
target = np.array(target)
print(target.shape)
```

```
from sklearn.utils import shuffle
X_train = final
Y_train = target

X_train,Y_train = shuffle(X_train,Y_train)
```

```
X_train = X_train/255.0
```

```
from tensorflow.keras.applications import InceptionResNetV2
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import InputLayer
from tensorflow.keras.layers import GlobalAveragePooling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Model
from tensorflow.keras import optimizers
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.layers import LSTM,Input,BatchNormalization
from tensorflow.keras.layers import TimeDistributed
```

```
import numpy as np

googleNet_model = InceptionResNetV2(include_top=False, weights='imagenet', input_shape=(128,128,3))
googleNet_model.trainable = True
cnn = Sequential()
cnn.add(googleNet_model)
cnn.add(GlobalAveragePooling2D())

rnn = Sequential()
rnn.add(LSTM(128,return_sequences=True,input_shape=(8,1536)))
rnn.add(LSTM(128,return_sequences=False))
rnn.add(Flatten())

dense = Sequential()
dense.add(Dense(64,activation = 'relu'))
dense.add(Dropout(0.5))
dense.add(Dense(5,activation='softmax'))




main_input = Input(shape=(8, 128, 128, 3)) # Data has been reshaped to (800, 5, 120, 60, 1)

model = TimeDistributed(cnn)(main_input) # this should make the cnn 'run' 5 times?
model = rnn(model) # combine timedistributed cnn with rnn
model = dense(model) # add dense
```

```
final_model = Model(inputs=main_input, outputs=model)
```

```
final_model.summary()
```

```
from keras.optimizers import Adam

Adam =optimizers.Adam(lr=0.0000001,beta_1=0.9, beta_2=0.999)
final_model.compile(loss="categorical_crossentropy",optimizer=Adam,metrics=['accuracy'])
```

```
final_model.fit(X_train,Y_train,validation_split=.2,batch_size=32,epochs = 100)
```
